### Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
from sklearn.preprocessing import RobustScaler
```

```python
param_dict_SVC = {'C': [0.1, 1, 10, 100],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf', 'linear', 'poly']}
```

```python
robust_scaler = RobustScaler()
Xtr_r = robust_scaler.fit_transform(X_train)
Xte_r = robust_scaler.transform(X_test)
```

```python
svc_classifier = SVC()
grid_search_SVC = GridSearchCV(svc_classifier, param_grid=param_dict_SVC, cv=5, verbose=1, n_jobs=-1)
grid_search_SVC.fit(Xtr_r, y_train)
```

```python
grid_search_SVC.best_params_
```

```python
svc_clf = SVC(C = 1, gamma = 1, kernel = 'linear')
svc_clf.fit(Xtr_r, y_train)
```

```python
#model evaluation and testing
y_pred_train = svc_clf.predict(Xtr_r) 
y_pred_test = svc_clf.predict(Xte_r)
```

```python
print("Results for training data: \n")
classification_report_train = classification_report(y_train, y_pred_train)
print("Classification Report for train data: \n", classification_report_train)

plot_roc_curve(svc_clf, Xtr_r, y_train)
```

```python
plot_confusion_matrix(svc_clf, Xtr_r, y_train)
```

```python
print("Results for testing data: \n")
classification_report_train = classification_report(y_test, y_pred_test)
print("Classification Report for test data: \n", classification_report_train)

plot_roc_curve(svc_clf, Xte_r, y_test)
```
